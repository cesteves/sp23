long double euler1 (int n) {
    long double sum = 0;
    long double fact;
    for (int i = 0; i <= n; i++) {
        fact = 1;
        for (int j = 2; j <= i; j++) {
            fact *= j; 
        }
        sum += (1.0/fact);
    }
    return sum;
}

uint64_t fibI(uint16_t n){
    uint64_t sum;
    uint64_t prev[] = {0,1};
    if (n < 2) return n;
    for (uint16_t i = 2; i <= n; i++) {
        sum = prev[0] + prev[1];
        prev[0] = prev[1];
        prev[1] = sum;
    }
    return sum;
}
